### **Array Operation**

***Arithmetic Operation:-***

```python
# the size of both array should be same for performing operation.
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])
```

```python
add=arr1+arr2
add
```

```python
sub=arr2-arr1
sub
```

```python
mul=arr1*arr2
mul
```

```python
div=arr2/arr1
div
```

```python
fl_div=arr2//arr1
fl_div
```

```python
exp=arr1**arr2
exp
```

***Broadcasting:-***

```python
l=[10,20,30,40,50]
arr=np.array(l)
```

```python
# If we want to add 10 in every element in list we cant write l+10 but we can arr+10
arr+10
```

```python
arr=np.arange(1,21).reshape(4,5)
arr
```

```python
arr+10
```

***Deep and Shallow Copy:-***

```python
a=np.arange(1,6)
a
```

```python
slice=a[1:3]
slice
# Here slice is creaing different array so a is not changing.
```

```python
b=a
b
```

```python
b[0]=99
b
```

```python
a
# Here a will change because of deep copy.
```

```python
# if we dont want to do this we can
b=a.copy()
b
```

### **Advance Array Manipulation:-**

***Stacking Array:-***

```python
A=np.array([[1,2,3,4]])
B=np.array([[5,6,7,8]])
```

```python
np.vstack((A,B))
```

```python
np.hstack((A,B))
```

```python
np.column_stack((A,B))
```

***Spliting Array:-***

```python
a=np.arange(1,17).reshape(4,4)
a
```

```python
np.hsplit(a,2)
```

```python
np.hsplit(a,4)
```

```python
np.vsplit(a,2)
```

```python
np.vsplit(a,4)
```
