### ***Numpy Indexing and Slicing of Vectors:-***

```python
arr=np.arange(1,11)
arr
```

```python
arr[5]
```

```python
# [2,3,4,5,6]
#arr[1:6]
arr[-9:-4]
```

### ***Numpy Indexing and Slicing of Matrix:-***

```python
arr=np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
# print 1 only
arr[0,0]
```

```python
# print 30 only
arr[5,4]
```

```python
# Print 
# 2  3
# 7  8
arr[0:2,1:3]
```

```python
# Print
#    19  20
#    24  25
#    29  30
arr[3:,3:]
```

```python
arr
```

```python
# Print
#    3
#    8
#    13
#    18
#    23
#    28
arr[0:,2]
```
